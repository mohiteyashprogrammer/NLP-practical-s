```python
from keras.datasets import imdb
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train.shape
```

```python
import numpy as np
sent = [
    "india loss worldcup",
    "virat played very well",
    "yash is good boy",
    "bharat mata ki jay",
    "dj wale babu mera gana chala do"
]
```

```python
# it is performing integer encoding
from keras.preprocessing.text import Tokenizer
tokens = Tokenizer()
```

```python
tokens.fit_on_texts(sent)
```

```python
tokens.word_index
```

```python
tokens.word_counts
```

```python
tokens.document_count
```

```python
encoding = tokens.texts_to_sequences(sent)
```

```python
encoding
```

```python
from keras.utils import pad_sequences
pad_sequences(encoding)
```

```python
pad_sequences(encoding,padding="post")
```

```python
X_train
```

```python
y_train
```

```python
X_train = pad_sequences(X_train,padding="post",maxlen=100)
```

```python
X_test = pad_sequences(X_test,padding="post",maxlen=100)
```

```python
model = Sequential()
```

```python
model.add(SimpleRNN(32,input_shape=(100,1)))
```

```python
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=8,batch_size=42)
```

```python
model.predict(X_test)
```

```python
model.evaluate(X_test,y_test)
```
