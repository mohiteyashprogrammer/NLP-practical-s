```python
import pandas as pd
import numpy as np
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
```

```python
# this function help to do tokenize the words and sentances
simple_preprocess("yash mohite is good boy")
```

```python
# Create an empty list to store the processed sentences
story = []

# Iterate through files in the "data" directory
for filename in os.listdir("data"):
    # Build the full path to the file
    file_path = os.path.join("data", filename)

    # Open the file with the specified encoding (e.g., 'ISO-8859-1')
    with open(file_path, 'r', encoding='ISO-8859-1') as f:
        # Read the content of the file
        corpus = f.read()

        # Tokenize the text into sentences
        raw_sent = sent_tokenize(corpus)

        # Process each sentence and append to the story list
        for sent in raw_sent:
            story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
# buigld model
model = gensim.models.Word2Vec(
    window = 10,
    min_count=2,
    workers=4
)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar("daenerys")
```

```python
model.wv.most_similar("jon")
```
