```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(u"I will google about facebook and instagram also about snapchat")
```

```python
doc.text
```

```python
doc[1]
```

```python
doc[5].pos_
```

```python
doc[3].tag_
```

```python
spacy.explain("IN")
```

```python
for word in doc:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc2 = nlp(u"I left the room")
for word in doc2:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc3 = nlp(u"to the left of the room")
for word in doc3:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```
